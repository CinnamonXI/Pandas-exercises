### Step 1. Import the necessary libraries

```python
import pandas as pd
```

### Step 2. Import the dataset from this [address](https://raw.githubusercontent.com/justmarkham/DAT8/master/data/chipotle.tsv). 

```python
data = pd.read_table("https://raw.githubusercontent.com/justmarkham/DAT8/master/data/chipotle.tsv")
```

### Step 3. Assign it to a variable called chipo.

```python
chipo = data
chipo.head()
```

### Step 4. How many products cost more than $10.00?

```python
chipo.info()
```

```python
floater = lambda x : float(x[1:-1])
chipo.item_price = chipo.item_price.apply(floater)
chipo.info()
```

```python
chipo_fil = chipo.drop_duplicates(["quantity", "item_name", "choice_description"])
chipo_fil
```

```python
chipo_fil[(chipo_fil.quantity == 1) & (chipo_fil.item_price > 10)]
```

```python
chipo_fil[(chipo_fil.quantity == 1) & (chipo_fil.item_price > 10)].item_name.nunique()
```

### Step 5. What is the price of each item? 
###### print a data frame with only two columns item_name and item_price

```python
chipo_price = chipo_fil[(chipo_fil.quantity == 1) & (chipo_fil.item_price > 10)]
del chipo_price["order_id"]
del chipo_price["quantity"]
del chipo_price["choice_description"]
chipo_price
```

### Step 6. Sort by the name of the item

```python
chipo_price.sort_values(by="item_name" ,ascending= True)
```

### Step 7. What was the quantity of the most expensive item ordered?

```python
chipo_fil.sort_values(by="item_price", ascending=False)
```

### Step 8. How many times was a Veggie Salad Bowl ordered?

```python
vsb = chipo[chipo.item_name == "Veggie Salad Bowl"]
vsb
```

```python
vsbc = vsb.value_counts().count()
vsbc
```

### Step 9. How many times did someone order more than one Canned Soda?

```python
cs = chipo[(chipo.item_name == "Canned Soda") & (chipo.quantity > 1)]
cs.value_counts().count()
```
